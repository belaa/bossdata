Graph RA,DEC,STO,and Illumination at Time (initially given in JD). Data from Horizons Web-Interface. 

Ephemeris: Observer, Target Body: Moon(Luna), Observer Location: Irvine,CA, Time Span: 2015,Table Setting: 1,4,10,24,angleformat=Deg

```python
%pylab inline
```

```python
from astropy.time import Time
import astropy.table as tab
import matplotlib.pyplot as plt
```

```python
data_year=tab.Table.read("/home/dianne/Downloads/horizons_results15.f.txt",format='ascii')
```

```python
print data_year
```

```python
Azi_Elevplot=plt.plot(data_year["col7"],data_year["col8"],':k', label='Azi-Elev')
plt.xlabel('Azimuth (measure N,E,S,W)')
plt.ylabel('Elevation (Topocentric degrees)')
plt.title('Azimuth and Elevation of Target Center')
plt.legend()
```

```python
Azi_timeplot=plt.plot(data_year["col3"],data_year["col8"],':b', label='Azi')
plt.xlabel('Azimuth (measure N,E,S,W)')
plt.ylabel('Time (JD)')
plt.title('Azimuth Changes of Moon in 2015')
plt.legend()
```

```python
Elev_timeplot=plt.plot(data_year["col3"],data_year["col7"],':g', label='Elev')
plt.xlabel('Elevation (Topocentric degrs to local zenith dir)')
plt.ylabel('Time (JD)')
plt.title('Elevation of Moon in 2015')
plt.legend()
```

```python
RA_Decplot=plt.plot(data_year["col5"],data_year["col6"],':c', label='RA-DEC')
plt.xlabel('Right Ascension (degrees-adjusted for time light)')
plt.ylabel('Declination ( degrees)')
plt.title('RA and DEC of Moon 2015')
plt.legend()
```

```python
RA_timeplot=plt.plot(data_year["col3"],data_year["col5"],':m', label='RA')
plt.xlabel('Right Ascension (degrs-adjusted for time light)')
plt.ylabel('Time (JD)')
plt.title('RA of Moon 2015')
plt.legend()
```

```python
Dec_timeplot=plt.plot(data_year["col3"],data_year["col6"],'c--', label='DEC')
plt.xlabel('Declination (degrees)')
plt.ylabel('Time (JD)')
plt.title('DEC of Moon 2015')
plt.legend()
```

```python
Illum_timeplot=plt.plot(data_year["col3"],data_year["col9"],'b--', label='Illum')
plt.xlabel('Observer Illumination (Percent)')
plt.ylabel('Time (JD)')
plt.title('Fraction of Moon Illuminated by Sun 2015')
plt.legend()
```

```python
SOTangle_time=plt.plot(data_year["col3"],data_year["col10"],':g', label='SOT')
plt.xlabel('S-T-O (Degrees)')
plt.ylabel('Time (JD)')
plt.title('Interior Vertex Angle of Moon (Sun Apparent Center and Observer)')
plt.legend()
```
